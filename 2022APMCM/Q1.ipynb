```python
import matplotlib.pyplot as plt

# 数据
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
average_monthly_temperatures = [0.784, 0.918, 1.062, 1.444, 2.204, 1.862, 1.392, 1.822, 1.767, 1.404, 1.536]
differences = [0, 0.134, 0.144, 0.382, 0.76, -0.342, -0.47, 0.43, -0.055, -0.363, 0.132]
plt.rc('font',size=16)
plt.rc('font',family='SimHei')

# 创建一个新的图像
plt.figure()

# 绘制月平均气温
plt.plot(years, average_monthly_temperatures, marker='o', label='3月平均气温')

# 绘制一阶差分
bars = plt.bar(years, differences)


# # 在每个柱上显示数值
# for bar in bars:
#     yval = bar.get_height()
#     plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, yval, ha='center', va='bottom')

# 添加图例
plt.legend()

# 添加x轴和y轴标签
plt.xlabel('年份')
plt.ylabel('值')

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [-0.75,-0.5,-0.25,0,0.25,0.5,0.75]
y = [1,2,1,3,1,1,1]
plt.rc('font', size=16)
plt.rc('font', family='SimHei')
plt.rcParams['axes.unicode_minus']=False  # 解决负号显示问题

plt.figure()
plt.bar(x, y, width=0.1, color='skyblue')  # 添加颜色和宽度
plt.legend()
plt.xlabel('一阶差分值')
plt.ylabel('频率')
plt.show()
```

```python
# 导入需要的库
import numpy as np 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# 加载iris鸢尾花数据集
iris = datasets.load_iris()

# 获取特征矩阵和目标向量
X = iris.data  
y = iris.target

# 划分训练数据集和测试数据集,test_size=0.3表示测试集大小为30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)  

# 定义KNN分类器,n_neighbors=3表示考虑3个最近邻样本
knn = KNeighborsClassifier(n_neighbors=3)

# 使用训练数据fit KNN模型
knn.fit(X_train, y_train) 

# 对测试集进行预测
y_pred = knn.predict(X_test)

# 计算预测准确率
accuracy = np.mean(y_pred == y_test) * 100
print("Accuracy:", accuracy)
```
